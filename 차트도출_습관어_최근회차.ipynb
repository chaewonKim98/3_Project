```python
f1 = open('/Users/SM315/Project/test_speak1.txt','r',encoding='euc-kr')
txt = f1.readlines()
f1.close()
```

```python
morph_list = []
for i in txt :
  morphs = okt.morphs(i)
  for morph in morphs:
    morph_list.append(morph)
```

```python
def counter(morph_list):
  word_count = {}
  for word in morph_list:
    if word in word_count :
      word_count[word] += 1
    else :
      word_count[word] = 1
  return word_count

word_count = counter(morph_list)
word_count
```

```python
# 정렬
word_count = sorted(word_count.items(), key = lambda x:x[1], reverse=True)
word_count
```

```python
fields = ['습관어' , '횟수']
rows = word_count
```

```python
with open(r'test1.csv', 'w', newline='\n') as f:
  write = csv.writer(f)
  write.writerow(fields)
  write.writerows(rows)
```

```python
for item in data1:
  if item =='아':
    cnt[0] = cnt[0] +1
  elif item == '음':
    cnt[1] = cnt[1] +1
  elif item == '그':
    cnt[2] = cnt[2] +1
  elif item == '어':  
    cnt[3] = cnt[3]+1
  elif item == '그니까':
    cnt[4] = cnt[4] +1
  elif item == '저':
    cnt[5] = cnt[5] +1
  elif item == '이제':
    cnt[6] = cnt[6] +1
  elif item == '막':
    cnt[7] = cnt[7] +1
  elif item == '에':
    cnt[8] = cnt[8] +1
print(cnt)    
```

```python
# 습관어 list
Habitual_word = ['아','음','그','어','그니까','저','이제','막','에']
```

```python
x = np.arange(len(Habitual_word))
x
```

```python
plt.figure(figsize=(10,5))
plt.rc('font', family='Malgun Gothic')
plt.xlim(-1,9)
plt.ylim(0,15)
plt.bar(x-0.1, Habitual_word_cnt, width =0.3, color ='yellow',align='center')
plt.title('습관어 빈도수 분석결과', fontsize = 20)
plt.xlabel('<습관어 종류>', fontsize = 14)
plt.ylabel('<빈도수>', fontsize = 14)
plt.rc('xtick', labelsize = 12)
plt.rc('ytick', labelsize = 12)
#plt.rcParams['figure.figuresize'] = (10,4)
plt.grid(axis='y', linestyle='--', alpha = 0.5)
plt.xticks(x, Habitual_word, fontsize=8)

for i, v in enumerate(x):
    aa = "{:,}회    ".format(Habitual_word_cnt[i])
    plt.text(v,Habitual_word_cnt[i],aa,fontsize=9,color='#997000',
              horizontalalignment='center',verticalalignment='bottom')

plt.savefig('C://습관어의 종류와 빈도수_최근회차.png',bbox_inches='tight')
```
