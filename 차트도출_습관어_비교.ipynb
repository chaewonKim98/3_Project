```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
f1 = open('/Users/SM315/Project/test_speak1.txt','r',encoding='euc-kr')
txt = f1.readlines()
f1.close()
```

```python
morph_list = []
for i in txt :
  morphs = okt.morphs(i)
  for morph in morphs:
    morph_list.append(morph)
```

```python
len(morph_list)
```

```python
def counter(morph_list):
  word_count = {}
  for word in morph_list:
    if word in word_count :
      word_count[word] += 1
    else :
      word_count[word] = 1
  return word_count

word_count = counter(morph_list)
word_count
```

```python
# 정렬
word_count = sorted(word_count.items(), key = lambda x:x[1], reverse=True)
word_count
```

```python
fields = ['습관어' , '횟수']
rows = word_count
```

```python
import csv
```

```python
with open(r'test1.csv', 'w', newline='\n') as f:
  write = csv.writer(f)
  write.writerow(fields)
  write.writerows(rows)
```
